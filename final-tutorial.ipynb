# 1. Data collection

Does population play a roll in GDP, let alone a counties GDP? 

The first step in exploring GDP by county is to look for public datasets, in this case we are accessing 2 datasets that gives us information on all the counties in the state of Maryland.

You can access these datasets through opendata.maryland.gov.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

gdp_table = pd.read_csv("https://opendata.maryland.gov/api/views/p5hr-8uyb/rows.csv?accessType=DOWNLOAD&bom=true&format=true")
pop_table = pd.read_csv("https://opendata.maryland.gov/api/views/r7ky-rq9s/rows.csv?accessType=DOWNLOAD&bom=true&format=true")
gdp_table
```

```python
pop_table
```

Using the built in function from pandas, we are able to read the csv files into dataframes as shown above. Its a little messy and everywhere right now, but we have successfully retreived and collected the data we need to explore this question.

# 2. Data management/representation

In order to know what we are even looking at we need to first clean up our two datasets from any noise or empty entries.
In this case both of our datasets have all fields populated, but there are still extra columns that are not needed.
What we want to look at are the columns of every county, and in these datasets, the provided years of gdp and population within the counties.

The extra data are what we need to drop.

```python
df_gdp = pd.DataFrame(gdp_table)
df_pop = pd.DataFrame(pop_table)

df_gdp = df_gdp.drop(df_gdp.columns[0], axis=1)
df_gdp = df_gdp.drop(df_gdp.columns[1], axis=1)
df_gdp = df_gdp.reset_index(drop=True)  

df_pop = df_pop.drop(df_pop.columns[0], axis=1)
df_pop = df_pop.drop(df_pop.columns[1], axis=1)
df_pop = df_pop.reset_index(drop=True)  

df_gdp = df_gdp.set_index('Year')
df_pop = df_pop.set_index('Year')

df_gdp.T

```

```python
df_pop.T
```

If you take a look at the above dataframes, you can see that the date created and maryland totals of gdp and population were removed.
Then you can see that the year column was converted into indices. The displayed dataframes are transposed of the original to better understand the data and view each year completely by county.
However, this is still slightly messy, there are no trends to be seen and there is still more room to visualize the data better.

# 3. Exploratory data analysis

From looking at the transpose datasets, what can be seen right away are the years and counties.

In this example, We've analyzed the years and found that there are a few overlapping years across the two datasets.
What we can do now is merge the two datasets by the years and since we are indexing by year we can just match up the years that overlap to creaty a merged dataframe. However, we have 2 different data, one looks at the gdp of a county, the other looks at the population.

One way arround this is to make two dataframes in respect of the datasets.

```python
df_merged_pop = pd.merge(df_pop, df_gdp, on=['Year'], how='inner')
df_merged_gdp = pd.merge(df_gdp, df_pop, on=['Year'], how='inner')
df_merged_gdp = df_merged_gdp.T
df_merged_pop = df_merged_pop.T

df_merged_gdp
```

```python
df_merged_pop
```

As seen above, we now have two datasets for gdp and population of each county, and whats more is we have narrowed the data to 4 years at which we will be looking and analyzing to see if we can depict any trends.
A thing to notice as well is that each dataset has repeated data half way down the rows, but we will address that when we progress.

A hypothesis we can state is; If a population is high, the gdp will be high.

So how can we test this with the datasets above?
One way we can do this is to make a relation of gdp and population and make a new column of the relation.

# 4. Hypothesis testing

Here we need to look at each county and merge them together by gdp and population. This took a little bit of trial and error, and some witty intuitive thinking but a solution that was found was to make 4 different dataframes and look at the data for each year.

A bump in the road when creating a third column for these 4 new dataframes was the calculation of the ratio. Because the data types were of type object, and the entries were in string, you can't do a simple astype() function to convert it into a float and make a ratio. 

First we needed to use a delimiter and remove unncessary punctuation from the numbers to allow us to cast them as float. From there we can finally input the result in the ratio column of the four dataframes created by the intersecting years of the two datasets that we are originally analyzing.
Then we converted each of the four datasets to its Transpose to visualize it better, however, another issue that we still needed to address was the doubling of counties. Since the counties were our rows we simply did a drop function to drop everything after the last county to have the only data we need. Making the four datasets into transpose also helps understand and figure out what graph(s) are best to explore our hypthoesis.

```python
import re
i = 1
for year in df_merged_gdp:
    if i == 1:
        df_yr2010 = pd.DataFrame([df_merged_gdp[year], df_merged_pop[year]])
        i+=1
    elif i == 2:
        df_yr2015 = pd.DataFrame([df_merged_gdp[year], df_merged_pop[year]])
        i+=1
    elif i == 3:
        df_yr2016 = pd.DataFrame([df_merged_gdp[year], df_merged_pop[year]])
        i+=1
    elif i == 4:
        df_yr2020 = pd.DataFrame([df_merged_gdp[year], df_merged_pop[year]])
        i+=1


df_yr2010.index = ['GDP', 'Population']
df_yr2015.index = ['GDP', 'Population']
df_yr2016.index = ['GDP', 'Population']
df_yr2020.index = ['GDP', 'Population']

#2010 ratio
df_yr2010 =  df_yr2010.T
df_yr2010 = df_yr2010.drop(df_yr2010 .index[24:])
df_yr2010['Ratio'] = \
    (df_yr2010['GDP'].str.replace(',','')).astype(float)/(df_yr2010['Population'].str.replace(',','')).astype(float)
    		
#2015 ratio
df_yr2015 =  df_yr2015.T
df_yr2015 = df_yr2015.drop(df_yr2015 .index[24:])
df_yr2015['Ratio'] = \
    (df_yr2015['GDP'].str.replace(',','')).astype(float)/(df_yr2015['Population'].str.replace(',','')).astype(float)
#2016 ratio
df_yr2016 =  df_yr2016.T
df_yr2016 = df_yr2016.drop(df_yr2016 .index[24:])
df_yr2016['Ratio'] = \
    (df_yr2016['GDP'].str.replace(',','')).astype(float)/(df_yr2016['Population'].str.replace(',','')).astype(float)
    
#2020 ratio
df_yr2020 =  df_yr2020.T
df_yr2020 = df_yr2020.drop(df_yr2020 .index[24:])
df_yr2020['Ratio'] = \
    (df_yr2020['GDP'].str.replace(',','')).astype(float)/(df_yr2020['Population'].str.replace(',','')).astype(float)
    
df_yr2010
    
```

Looking at the four datasets above we can see that we now have our ratio column and 24 county indices. What we want to do is look at the ratios.
In this step, what was seen as the best course of action was to use a horizontal bar graph. This way we can have a visual representation of the differences between each ratio of their counties to other counties. Four bar graphs were graphed to see if there were and visual differences.

```python
idx = 0

#2010 ratio
ax1 = plt.subplot(1,1,1)
ax1.barh(df_yr2010.index, df_yr2010['Ratio'])
ax1.set_xlim(0, df_yr2010['Ratio'].values.max() * 1.2)
ax1.set_ymargin(.0125)
ax1.xaxis.grid(linestyle='--', linewidth = .5)
ax1.set_axisbelow(True)
ax1.set_xlabel('gdp/county-population: %')
ax1.set_title('GDP county population ratio in 2010', fontsize = 15, weight = 'bold')
for Y, X in enumerate(df_yr2010['Ratio'].values):
    ax1.annotate('{:%}'.format(X), xy= (X,Y))
plt.show()

#2015 ratio
ax2 = plt.subplot(1,1,1)
ax2.barh(df_yr2015.index, df_yr2015['Ratio'])
ax2.set_xlim(0, df_yr2015['Ratio'].values.max() * 1.2)
ax2.set_ymargin(.0125)
ax2.xaxis.grid(linestyle='--', linewidth = .5)
ax2.set_axisbelow(True)
ax2.set_xlabel('gdp/county-population: %')
ax2.set_title('GDP county population ratio in 2015', fontsize = 15, weight = 'bold')
for Y, X in enumerate(df_yr2015['Ratio'].values):
    ax2.annotate('{:%}'.format(X), xy= (X,Y))
plt.show()

#2016 ratio
ax3 = plt.subplot(1,1,1)
ax3.barh(df_yr2016.index, df_yr2016['Ratio'])
ax3.set_xlim(0, df_yr2016['Ratio'].values.max() * 1.2)
ax3.set_ymargin(.0125)
ax3.xaxis.grid(linestyle='--', linewidth = .5)
ax3.set_axisbelow(True)
ax3.set_xlabel('gdp/county-population: %')
ax3.set_title('GDP county population ratio in 2016', fontsize = 15, weight = 'bold')
for Y, X in enumerate(df_yr2016['Ratio'].values):
    ax3.annotate('{:%}'.format(X), xy= (X,Y))
plt.show()

#2020 ratio
ax4 = plt.subplot(1,1,1)
ax4.barh(df_yr2020.index, df_yr2020['Ratio'])
ax4.set_xlim(0, df_yr2020['Ratio'].values.max() * 1.2)
ax4.set_ymargin(.0125)
ax4.xaxis.grid(linestyle='--', linewidth = .5)
ax4.set_axisbelow(True)
ax4.set_xlabel('gdp/county-population: %')
ax4.set_title('GDP to county population ratio in 2020', fontsize = 15, weight = 'bold')
for Y, X in enumerate(df_yr2020['Ratio'].values):
    ax4.annotate('{:%}'.format(X), xy= (X,Y))

plt.show()
```

### Extended thinking

Although we have plotted the relations of population and gdp, this data holds many external factors that become variables in changing the data making reading the ratio difficult to interpret.
As stated previously, Kent County theoretically denotes a relationship of such where an increase in the GDP will increase the GDP over population by a factor of almost 3. However, realistically, other factors such as physical size, county poverty itself, and markets available within the county can affect this ratio.

To give this some perspective, Baltimore County, Maryland’s most population heavy county, has only a ratio of 0.06. As per why this is so small, consider the aspect of poverty. From a wtop news article, we can see that Baltimore City within the county has the highest poverty rate, at a staggering 21.8% of Maryland, while Kent is only at 11% (Valerie Bonk).
